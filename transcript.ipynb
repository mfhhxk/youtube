# Download Youtube video
Download the given `video_url` from YouTube as audio-only and save it locallaly.

```python
from pytube import YouTube

video_url = "https://www.youtube.com/watch?v=fA_yDjx27NA&ab_channel=%E8%A9%A6%E7%95%B6%E7%9C%9FTrial%26Error" #@param {type:"string"}
audio_file = YouTube(video_url).streams.filter(only_audio=True).first().download(filename="audio.mp4")
```

# Load the Whisper Model

```python
import whisper

whisper_model = whisper.load_model('medium')
```

# Transcript the Video

```python
transcription = whisper_model.transcribe(audio_file)
```

# Display Transcription as DF

```python
import pandas as pd
df = pd.DataFrame(transcription['segments'], columns=['start', 'end', 'text'])
display(df)
```

# Sentiment Analysis on the shorter video

```python
result = whisper_model.transcribe("video.mp4", verbose = True)
print(result["text"])
```

```python
for i, seg in enumerate(result['segments']):
    print(i+1, "- ", seg['text'])
```

```python
result
```

```python
import pandas as pd
# store in dataframe for later sentiment analysis
speech = pd.DataFrame.from_dict(result['segments'])
speech.head()
```

```python
# filter the text column and remove the remaining 
speech = speech[['text']] 

# randomly select 20 segments from the dataframe 
speech = speech.sample(n = 20) 
speech.head()
```

```python
# !pip install ipywidgets
# !jupyter nbextension enable --py widgetsnbextension
```

Model used: https://huggingface.co/lxyuan/distilbert-base-multilingual-cased-sentiments-student  
support Multilingual Sentiment analysis including Chinese

```python
from transformers import pipeline

classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student", 
    return_all_scores=True
)
```

```python
speech.text
```

```python
def reshape(row):
    lst = classifier(row)[0]
    tmp = dict()
    tmp['pos'] = lst[0].get('score')
    tmp['neu'] = lst[1].get('score')
    tmp['neg'] = lst[2].get('score')
    return tmp
    
```

ref: https://www.youtube.com/watch?v=QpzMWQvxXWk 

```python
reshape_dict = {}
for i,row in enumerate(speech['text']):
    reshape_dict[i] = reshape(row)
```

```python
reshape_dict
```

```python
# Merge with speech
reshape_df = pd.DataFrame(reshape_dict).T
reshape_df = reshape_df.reset_index().rename(columns={'index':'id'})
reshape_df = reshape_df.merge(speech, how='left').set_index('id')
```
